```python
import numpy as np
import math
import random
```

```python
# Genarate Data
x = np.random.randint(500, size=1000)
x = np.array([1,2,3,5,4,6,6,1,2,5,5,6,8,9,1,2,7,8,10,23,10])
x = np.sort(x)
print(x)
```

```python
# get f(x) of normalization distribute
mean = x.mean()
sd = x.std()

# Is this x_range correct?
x_nomal = np.array(range(int(mean-3*sd-1), int(mean+3*sd+1)))

print("mean : {0}, standard deviation : {1}".format(mean,sd))
y = np.exp(-((x_nomal-mean)**2)/(2*sd**2))/(sd*(2*math.pi)**(1/2))
```

```python
Z7 = (7-mean)/sd
print(Z7) #0.42 percent
print(6/20*100)
```

```python
# 95% of value between -2sd , 2sd
print("95% on [{},{}]".format(mean-2*sd, mean+2*sd))
```

```python
# plot graph
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

ax.plot(x_nomal, y)

# mean +- 1 sd 
ax.plot([mean for i in range(len(x_nomal))],y, label=str(mean))

plt.legend()
plt.show()
```

```python
# convert to z-score
z = (x_nomal-mean)/sd 
y = np.exp(-1*(z**2)/(2))/((2*math.pi)**(1/2))
```

```python
# plot graph
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(z, y)
plt.show()
```
